```python
'''
In this tutorial we will make a custom plot of the RTMA Analysis for temperature across the central U.S.
We will want to use a custom area and a custom reference system. 
When making custom plots, there is a lot more "playing around with" the optional arguments inside of the function. 
When making a plot of a preset state/gacc region and using a preset reference system, there is much less optional
arguments needing to be modified. 

Now let's begin our first step and import the RTMA module from firewxpy
'''
from firewxpy import rtma_conus as rtma
```

```python
'''
To create a completely custom plot the following variables need to be set to the following values:

1) state=None (Default: state='us')
2) gacc_region=None (Default: gacc_region=None)
3) reference_system='custom'

Then the user must enter the latitude/longitude coordinates for the bounds of the plot. 

The user then must determine the following:
1) figure size (fig_x_length, fig_y_length)
2) title fontsize
3) Analysis timestep fontsize (subplot_title_fontsize)
4) Position and fontsize of the signature textbox
5) The types of borders the user wishes to overlay
    - In this example we overlay the following borders:
        1) state borders
        2) GACC borders
        3) NWS Public Zone borders

Lastly, you will see the folders generated in the new branch of the directory titled "Custom"
That is because the state and/or gacc_region as well as the reference_system names are part 
of the path to the images. In this case, rather than a state or gacc_region abbreviation, the 
folder is Custom since it is based on a custom area. 

'''

rtma.plot_temperature(western_bound=-110, eastern_bound=-80, southern_bound=30, northern_bound=51, fig_x_length=9, fig_y_length=9, signature_x_position=0.01, signature_y_position=0.1, color_table_shrink=1, title_fontsize=12, subplot_title_fontsize=10, signature_fontsize=8, colorbar_fontsize=8, show_rivers=True, reference_system='Custom', show_state_borders=True, show_county_borders=False, show_gacc_borders=True, show_psa_borders=False, show_cwa_borders=False, show_nws_firewx_zones=False, show_nws_public_zones=True, state=None, gacc_region=None)
```
